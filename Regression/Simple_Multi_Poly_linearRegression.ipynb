```python
#We should remove the feature "Balcony" "Status" "parking" "Furnished_status" "Lift" "Landmarks"

# List of columns to drop
columns_to_drop = ["Balcony", "Status", "parking", "Furnished_status", "Lift", "Landmarks"]

delhi.drop(columns=columns_to_drop,inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

def simple_linear_regression(X, y):
    """
    Performs simple linear regression on the provided data and calculates the accuracy.

    Parameters:
    X (array-like): 1D array of predictor variable (independent variable).
    y (array-like): 1D array of response variable (dependent variable).

    Returns:
    model (LinearRegression): The fitted linear regression model.
    r_squared (float): The R-squared value indicating the goodness of fit.
    mse (float): The mean squared error of the model.
    """
    # Ensure X is a 2D array (needed for scikit-learn)
    X = np.array(X).reshape(-1, 1)
    y = np.array(y)

    # Create and fit the model
    model = LinearRegression()
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Calculate accuracy metrics
    r_squared = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)

    return model, r_squared, mse
```

```python
X=delhi["area"]
y=delhi["price"]
model, r_squared, mse = simple_linear_regression(X, y)
print(f"R-squared: {r_squared}")
print(f"Mean Squared Error: {mse}")
print(f"Slope: {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")
```

```python
def multiple_linear_regression(X, y):
    """
    Performs multiple linear regression on the provided data and calculates the accuracy.

    Parameters:
    X (array-like): 2D array of predictor variables (independent variables).
    y (array-like): 1D array of response variable (dependent variable).

    Returns:
    model (LinearRegression): The fitted linear regression model.
    r_squared (float): The R-squared value indicating the goodness of fit.
    mse (float): The mean squared error of the model.
    """
    # Ensure X is a 2D array (needed for scikit-learn)
    X = np.array(X)
    y = np.array(y)

    # Create and fit the model
    model = LinearRegression()
    model.fit(X, y)

    # Make predictions
    y_pred = model.predict(X)

    # Calculate accuracy metrics
    r_squared = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)

    return model, r_squared, mse
```

```python
X=delhi[["area","Bedrooms","Bathrooms","Price_sqft"]]
y=delhi["price"]
model, r_squared, mse = multiple_linear_regression(X, y)
print(f"R-squared: {r_squared}")
print(f"Mean Squared Error: {mse}")
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
from sklearn.preprocessing import PolynomialFeatures

def polynomial_regression(X, y, degree=2):
    """
    Perform polynomial regression on the provided data and calculate the accuracy.

    Parameters:
    X (array-like): 2D array of predictor variables (independent variables).
    y (array-like): 1D array of response variable (dependent variable).
    degree (int): The degree of the polynomial features.

    Returns:
    model (LinearRegression): The fitted polynomial regression model.
    r_squared (float): The R-squared value indicating the goodness of fit.
    mse (float): The mean squared error of the model.
    """
    # Transform features for polynomial regression
    poly = PolynomialFeatures(degree=degree, include_bias=False)
    X_poly = poly.fit_transform(X)

    # Create and fit the model
    model = LinearRegression()
    model.fit(X_poly, y)

    # Make predictions
    y_pred = model.predict(X_poly)

    # Calculate accuracy metrics
    r_squared = r2_score(y, y_pred)
    mse = mean_squared_error(y, y_pred)

    return model, r_squared, mse
```

```python
X=delhi[["area","Bedrooms","Bathrooms","Price_sqft"]]
y=delhi["price"]

# Perform polynomial regression
model, r_squared, mse = polynomial_regression(X, y, degree=2)

# Output the results
print(f"R-squared: {r_squared}")
print(f"Mean Squared Error: {mse}")
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```
